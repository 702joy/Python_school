```python
import csv
f=open('weather.csv',encoding='cp949')
data=csv.reader(f)
header=next(data)
data=list(data)
f.close()
```

```python
print(len(data))
```

```python
for row in data[:3]:
  print(row[0])
```

```python
avg=[]
years=[1960]
sum=0
days=0
for row in data:
  if int(row[0].split('-')[0])==years[-1]:
    sum+=float(row[1])
    days+=1
  elif int(row[0].split('-')[0])>years[-1]:
    avg.append(sum/days)
    if int(row[0].split('-')[0])==2021:
      break 
    else:
      years.append(years[-1]+1)
      sum = float(row[1])
      days = 1
plt.figure(figsize=(15,4))
plt.xlabel('연도')
plt.ylabel('기온')
plt.plot(years,avg,'red')
plt.title('1960~2020년의 평균기온 그래프')
plt.legend()
plt.show()
```

```python
result=[]

for row in data:
  if row[2] != '':
    result.append(float(row[2]))
plt.hist(result,bins=100)
plt.show()
```

```python
aug = []
jan = []
for row in data:
  month=row[0].split('-')[1]
  if row[2] != '':
    if month == '08':
      aug.append(float(row[2]))
    if month == '01':
      jan.append(float(row[2]))
plt.hist(aug,bins=100,color='red',label='aug')
plt.hist(jan,bins=100,color='blue',label='jan')
plt.legend()
plt.show()
```
