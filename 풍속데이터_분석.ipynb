```python
import csv
f = open('wind.csv',encoding='cp949')
data = csv.reader(f)
header = next(data)
data=list(data)
f.close
```

```python
for row in data[:3]:
  print(row)
```

```python
max=0
day=''
for row in data:
  if row[4] != '':
    if max<float(row[4]):
      max=float(row[4])
      day=row[0]
print(f'{len(data)}일 중에{day}에 최대순간풍속이  {sum}로 가장 컸다')
```

```python
child=[]
liberty=[]

for row in data:
  if row[0].split('-')[1]=='05' and row[0].split('-')[2]=='05':
    if row[1]!='':
      child.append(float(row[1]))
  if row[0].split('-')[1]=='08' and row[0].split('-')[2]=='15':
    if row[1]!='':
      liberty.append(float(row[1]))

plt.figure(figsize=(12,3))
year=list(range(1981,2022))
plt.plot(year,child,'skyblue',label='어린이날',marker='^',ms='4')
plt.plot(year,liberty,'sandybrown',label='광복절',marker='o',ms='4')
plt.title('어린이날과 광복절의 평균 풍속', fontsize=15)
plt.legend()
plt.xlabel('연도',fontsize=13)
plt.ylabel('평균 풍속',fontsize=13)
plt.show()
```

```python
max_wind=[]

for row in data:
  if row[3] !='':
    max_wind.append(float(row[3]))

plt.hist(max_wind,bins=30,color='skyblue')
plt.xlabel('최대풍속풍향')
plt.ylabel('빈도')
plt.show()
```

```python
month=[[]for i in range(12)]

for row in data:
  index=int(row[0].split('-')[1])-1
  if row[1] != '':
    month[index].append(float(row[1]))
plt.figure(figsize=(12,5))
plt.boxplot(month,showfliers=False)
plt.xlabel("월")
plt.ylabel("평균풍속")
plt.show()
```
